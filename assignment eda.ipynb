```python
# QUESTION.1 What are the key features of the wine quality data set? Discuss the importance of each feature in
# predicting the quality of wine.
# ANSWER Certainly! Let’s delve into the key features of the wine quality dataset and their significance in predicting
# wine quality.

# Fixed Acidity: This feature represents the concentration of non-volatile acids in the wine. It contributes to the
# overall taste and balance. Wines with higher fixed acidity tend to have a crisper, more refreshing taste.

# Volatile Acidity: Volatile acids are responsible for the vinegar-like smell in wine. While a small amount is desirable
# for complexity, excessive volatile acidity negatively impacts wine quality.

# Citric Acid: Citric acid adds freshness and enhances the fruity flavors in wine. It plays a crucial role in balancing
# sweetness and acidity.

# Residual Sugar: This feature indicates the amount of sugar left after fermentation. Higher residual sugar levels result
# in sweeter wines, while lower levels lead to drier wines.

# Chlorides: Chlorides affect the wine’s saltiness. Optimal levels contribute to a well-rounded taste, but excessive
# chlorides can be unpleasant.

# Free Sulfur Dioxide: This compound acts as a preservative and antioxidant. Adequate levels prevent wine oxidation and
# maintain freshness.

# Total Sulfur Dioxide: The sum of free and bound sulfur dioxide. It impacts wine stability and shelf life.

# Density: Density reflects the wine’s overall composition. It correlates with alcohol content and residual sugar.

# pH: pH influences taste perception. Wines with balanced pH (around 3.3 to 3.6) are more pleasant.

# Sulphates: Sulphates enhance wine aroma and prevent bacterial growth. They positively impact wine quality.

# Alcohol: Alcohol content significantly affects wine body, mouthfeel, and flavor. Higher alcohol levels often indicate
# better quality.

# Quality (Target): The wine quality score (ranging from 3 to 9) is the target variable. It combines various sensory 
# attributes, making it a critical feature for prediction.

# In summary, understanding these features allows us to build accurate models for wine quality prediction. Each feature
# contributes uniquely, and their interplay defines the overall wine experience. 
```

```python
# QUESTION.2 How did you handle missing data in the wine quality data set during the feature engineering process?
# Discuss the advantages and disadvantages of different imputation techniques.

# ANSWER Handling missing data during feature engineering is crucial for building accurate machine learning models.
# Let’s delve into the strategies for imputing missing values and their pros and cons:
# 1. No Missing Values:
# * Advantage:
#       If your dataset has no missing values, you can proceed directly with feature engineering and modeling.
# * Disadvantage:
#        Rarely encountered in real-world scenarios; most datasets have some missing data.

# 2. Deletion (Dropping Rows or Columns):
# * Advantages:
#       Simple and straightforward.
#       Removes problematic observations.
# * Disadvantages:
#       Loss of information.
#       May lead to biased results if missingness is not random.
#       Reduces the sample size.
# 3. Mean/Median Imputation:
# * Advantages:
#       Easy to implement.
#       Preserves the overall distribution.
# * Disadvantages:
#      Ignores relationships between features.
#      Increases variance if missingness is not random.
#      May distort correlations.
# 4. Mode Imputation (for Categorical Variables):
# * Advantages:
#     Suitable for categorical features.
#     Simple and quick.
# Disadvantages:
#     Ignores feature interactions.
#     May introduce bias if mode is overrepresented.
# 5. Regression Imputation:
# Advantages:
#    Utilizes relationships between features.
#    More accurate than mean/median imputation.
# * Disadvantages:
#    Requires a predictive model (e.g., linear regression).
#    Sensitive to model assumptions.
# 6. K-Nearest Neighbors (KNN) Imputation:
# * Advantages:
#    Considers feature similarity.
#    Handles both numerical and categorical data.
# * Disadvantages:
#    Computationally expensive.
#    Choice of k affects imputation quality.
# 7. Multiple Imputation:
# * Advantages:
#    Generates multiple imputed datasets.
#    Accounts for uncertainty.
# * Disadvantages:
#    Complex and resource-intensive.
#    Requires specialized software.
# 8. Domain-Specific Imputation:
# * Advantages:
#    Customized to the problem domain.
#    Incorporates expert knowledge.
# * Disadvantages:
#    Requires domain expertise.
#    May not generalize well.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create a displot (distribution plot) for each feature
for feature in ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
'pH', 'sulphates', 'alcohol']:
    sns.displot(df, x=feature, kde=True, rug=True, bins=20)
    plt.title(f"Distribution of {feature}")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()
# If it resembles a bell-shaped curve, it suggests that the data may follow a normal distribution.
# If it deviates significantly from a bell shape, it indicates non-normality.
```

```python
# QUESTION.6 Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?
# ANSWER 
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
wine_data = pd.read_csv('redwine.csv')

# Separate features and target variable
X = wine_data.drop('quality', axis=1)  # Replace 'target_column' with the actual target column name

# Standardize the data
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Initialize PCA with the desired number of components (e.g., all features initially)
pca = PCA(n_components=None)

# Fit PCA on the standardized data
pca.fit(X_standardized)

# Find the cumulative explained variance
explained_variance_ratio_cumsum = pca.explained_variance_ratio_.cumsum()

# Determine the number of components for 90% variance
num_components_90_var = (explained_variance_ratio_cumsum >= 0.9).argmax() + 1
num_components_90_var
```
